```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d

# Fixing random state for reproducibility
np.random.seed(19680801)
```

```python
def randrange(n, vmin, vmax):
    """
    Helper function to make an array of random numbers having shape (n, )
    with each number distributed Uniform(vmin, vmax).
    """
    return (vmax - vmin)*np.random.rand(n) + vmin
```

```python
fig = plt.figure(figsize=(10,10)) 
ax = fig.add_subplot(projection='3d')
#sphere
u, v = np.mgrid[0:2*np.pi:50j, 0:np.pi:50j]
x = np.cos(u)*np.sin(v)
y = np.sin(u)*np.sin(v)
z = np.cos(v)
# alpha controls opacity
ax.plot_surface(x, y, z, color="g", alpha=0.1)

#points
u, v = np.mgrid[0:2*np.pi:9j,-np.pi/2:np.pi/2:5j]
x = np.cos(u)*np.cos(v)
y = np.sin(u)*np.cos(v)
z = np.sin(v)
ax.scatter(x,y,z)

#wireframe
a = np.linspace(0, 2*np.pi, 30)
for v in np.linspace(-np.pi/2,np.pi/2,5):
    x = np.cos(a)*np.cos(v)
    y = np.sin(a)*np.cos(v)
    z = np.sin(v)
    ax.plot3D(x,y,z, alpha=0.3)
    
a = np.linspace(0, 2*np.pi, 30)
y = np.sin(a)
z = np.cos(a)
x = np.zeros(a.shape)
#ax.plot3D(x,y,z, alpha=0.3)
plt.show()
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
 
fig = plt.figure()
 
# syntax for 3-D projection
ax = plt.axes(projection ='3d')
 
# defining all 3 axes
z = np.linspace(0, 2*np.pi, 30)
x = np.sin(z)
y = np.cos(z)
 
# plotting
ax.plot3D(x, y, np.zeros(z.shape), 'green')
ax.set_title('3D line plot geeks for geeks')
plt.show()
```
