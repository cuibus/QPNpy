```python
class trajectory:
    def __init__(self):
        self.posx=[]
        self.posy=[]
    def addPoint(self, x,y): 
        self.posx.append(x)
        self.posy.append(y)
```

```python
def getTraps():
    return [3,2,4],[3,4,2]
def getTrajectory():
    return [2,3,4,5,5,5,5,],[1,1,1,1,2,3,4,]
def getStartEnd():
    return [2,4],[1,5]
def correct(arg):
    offset = -0.45*np.ones(len(arg[0]))
    return offset+arg[0],offset+arg[1]
```

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style

fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(projection='3d')

trapx, trapy = correct(getTraps())
ax1.bar3d(trapx, trapy, np.zeros(len(trapx)), 0.9*np.ones(len(trapx)), 0.9*np.ones(len(trapx)), np.ones(len(trapx)), color='orange',zorder=1)

tx,ty = correct(getTrajectory())
tx = tx[1:len(tx)-1] #start and end points are ommited
ty = ty[1:len(ty)-1]
ax1.bar3d(tx, ty, np.zeros(len(tx)), 0.9*np.ones(len(tx)), 0.9*np.ones(len(tx)), np.ones(len(tx)), color="blue",zorder=2.5)
for i in range(0,len(tx)):
    ax1.text(tx[i]+0.5, ty[i]+0.3, 1, i+1, color='red',zorder=10, weight='bold')
    
sx, sy = correct(getStartEnd())
ax1.bar3d(sx, sy, np.zeros(len(sx)), 0.9*np.ones(len(sx)), 0.9*np.ones(len(sx)), np.ones(len(sx)), color='cyan',zorder=10)
ax1.text(sx[0]+0.5, sy[0]+0.3, 1, 0, color='red',zorder=10, weight='bold')
ax1.text(sx[1]+0.5, sy[1]+0.3, 1, len(tx)+1, color='red',zorder=10, weight='bold')

ax1.set_xlabel('x axis')
ax1.set_ylabel('y axis')
ax1.set_zlabel('probability')

ax1.set_xlim([-0.5, 7.5])
ax1.set_ylim([-0.5, 7.5])
ax1.set_zlim([0,1.5])
ax1.set_zticks([0,0.5,1])
ax1.set_box_aspect(aspect = (3,3,1))
plt.show()
```
